```python
def read_pdf_text(resume_path):
    """
    Extracts text from a PDF file.

    Args:
        resume_path (file_path): The PDF file to be read.

    Returns:
        str: The extracted text from the PDF.
    """
    pdf_reader = PyPDF2.PdfReader(resume_path)
    text = ""
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        text += page.extract_text().strip()
    return text
```

```python
import os

def save_to_markdown(texts, directory):
    """
    Save each text in the `texts` list to a separate markdown file in the specified `directory`.
    
    Args:
    - texts (list of dict): A list of dictionaries, where each dictionary has 'filename' and 'content' keys.
    - directory (str): The directory where the markdown files will be saved.
    """
    # Create the directory if it doesn't exist
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    for text in texts:
        # Create the full path for each markdown file
        file_path = os.path.join(directory, text['filename'])
        
        # Write content to the markdown file
        with open(file_path, 'w', encoding='utf-8') as f:
            f.write(text['content'])
    
    print(f"Files saved successfully in {directory}")
```

```python
# New code with resume_JD_match and markdown saving
def process_candidates_and_save(candidates, job_description, llm_jd_match, output_directory):
    """
    Process the list of candidates and save their resume match output to markdown files.

    Args:
    - candidates (list): List of candidate details to process.
    - job_description (str): The job description for matching.
    - llm_jd_match: The function that processes each candidate against the job description.
    - output_directory (str): Directory to save the output markdown files.
    """
    markdown_files = []

    for i, candidate in enumerate(candidates, 1):
        # Invoke the LLM matching function
        resume_JD_match = llm_jd_match.invoke({"candidate": candidate, "job_description": job_description})
        
        # Extract the content
        content = resume_JD_match.content
        
        # Create a filename for each candidate
        filename = f'candidate_match_{i}.md'
        
        # Add to markdown_files list
        markdown_files.append({
            'filename': filename,
            'content': content
        })
    
    # Save the markdown files
    save_to_markdown(markdown_files, output_directory)
```

```python
candidates = [candidate_1, candidate_2]
job_description = job_description
output_directory = 'resume_match_reports'

# Invoke the function to process and save files
process_candidates_and_save(candidates, job_description, llm_jd_match, output_directory)
```
